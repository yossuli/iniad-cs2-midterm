# スタックとキュー

## print_process 関数

`processes`に操作を`''' 操作 '''`のように入れ、スタック、キューのデフォルトの値、それぞれの操作を`引数=`の形で与えると実行ごとにスタック、キューの中身が表示される


```python
from collections import deque


def return_diff(current, new):
    diff = list(set(current) ^ set(new))
    if len(diff) == 0:
        return new
    if len(current) > len(new):
        return f"{current} --> {diff[0]}\n実行後:{new}"
    return f"{current} <-- {diff[0]}\n実行後:{new}"


def print_process(
    processes: str,
    stack_default: list = [],
    que_default: deque = deque(),
    スタックにプッシュ: str | None = None,
    スタックからポップ: str | None = None,
    キューに挿入: str | None = None,
    キューから取り出す: str | None = None,
    キューに逆から挿入: str | None = None,
    キューから逆から取り出す: str | None = None,
):
    stack = stack_default
    que = que_default
    keys = [
        スタックにプッシュ,
        スタックからポップ,
        キューに挿入,
        キューから取り出す,
        キューに逆から挿入,
        キューから逆から取り出す,
    ]
    funcs = [
        stack.append,
        stack.pop,
        que.append,
        que.popleft,
        que.appendleft,
        que.pop,
    ]
    dict = {key: func for key, func in zip(keys, funcs) if key is not None}
    process_list = deque(
        processes.replace("(", " ")
        .replace(")", " ")
        .replace('"', "")
        .replace("'", "")
        .split()
    )
    loop = 1
    while len(process_list) > 0:
        tmp_stack = stack.copy()
        tmp_que = que.copy()
        process = process_list.popleft()
        if process in [key for i, key in enumerate(keys) if i % 2 == 1]:
            dict[process]()
        else:
            value = process_list.popleft()
            if value in [key for i, key in enumerate(keys) if i % 2 == 1]:
                dict[process](dict[value]())
            else:
                dict[process](value)
        print(loop, "stack", return_diff(tmp_stack, stack))
        print(loop, "deq", return_diff(list(tmp_que), list(que)))
        print("---")
        loop += 1
```

## 使用例


```python
print_process(
    """
Dequeue
Enqueue 6
Enqueue 2
Dequeue
Enqueue 9
Dequeue
Enqueue 3
Dequeue
Enqueue 8
Dequeue
Dequeue
""",
    que_default=deque([5, 4, 1]),
    キューから取り出す="Dequeue",
    キューに挿入="Enqueue",
)
```
